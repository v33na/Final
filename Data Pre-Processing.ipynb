```python
import numpy as np
import csv
import json
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
import matplotlib.pyplot as mlp
import pandas as pd
from pandas import DataFrame

from sqlalchemy import create_engine
```

```python
df = pd.read_csv('spotifys-worldwide-daily-song-ranking/data.csv')
df = df.drop('URL', axis=1)
df['Artist_and_Track_Name'] = df['Artist']+ ' - ' + df['Track Name']
# df.head()
df['Date'] = pd.to_datetime(df['Date'])
# df.tail()
filtered_df = df[(df['Date'] > '2016-12-31') & (df['Date'] < '2018-01-01')]
filtered_df
```

```python
# groupby artist and track then stream count
global_streams = filtered_df.groupby(['Artist_and_Track_Name'])['Streams'].sum().reset_index()

# # # sort descending order on stream count
global_df = global_streams.sort_values(['Streams'], ascending=False)
global_df.rename(columns = {'Streams':'All_Streams'}, inplace = True)
# # # create column "position"
global_df['Position'] = range(1, 1+len(global_df))
# # # create index to value to find top 10, and delete the rest
# # indexNames = global_df[ (global_df['Position'] >= 50) ].index
# # global_df.drop(indexNames , inplace=True)
# # # copy data
global_df.set_index('Position')

export_csv = global_df.to_csv('spotify_streams.csv', index=False)
spotify_streams = pd.read_csv('spotify_streams.csv', index_col=2)

spotify = pd.DataFrame(spotify_streams)
spotify
```

```python
features_df = pd.read_csv('top-spotify-tracks-of-2017/features_top_2017.csv')
features_df['Artist_and_Track_Name'] = features_df['artists']+ ' - ' + features_df['name']
features_df
```

```python
NEW_df = pd.merge(spotify, features_df, how='inner')
NEW_df = NEW_df.drop(['id', 'name', 'artists'], axis=1)
NEW_df['Position'] = range(1, 1+len(NEW_df))
ML_df = NEW_df.set_index('Position')
ML_df
```

```python
export_csv = ML_df.to_csv('spotify_ML.csv')
spotify_ML = pd.read_csv('spotify_ML.csv')
spotify_ML
```
